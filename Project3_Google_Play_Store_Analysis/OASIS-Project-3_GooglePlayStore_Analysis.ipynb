```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("apps.csv")
```

```python
df.head()
```

```python
df = df.drop("Unnamed: 0", axis=1)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.describe()
```

```python
df["Category"].value_counts().head(10)
```

```python
df["Category"].value_counts().head(10).plot(kind="bar")

plt.title("Top App Categories")
plt.xlabel("Category")
plt.ylabel("Number of Apps")

plt.show()
```

```python
df["Type"].value_counts()
```

```python
df["Type"].value_counts().plot(kind="pie", autopct="%1.1f%%")

plt.title("Free vs Paid Apps")

plt.show()
```

```python
plt.hist(df["Rating"])

plt.title("Ratings Distribution")
plt.xlabel("Ratings")
plt.ylabel("Frequency")

plt.show()
```

# Conclusion

- Most apps on the Google Play Store belong to the FAMILY and GAME categories.
- The majority of apps are free, while paid apps are very few.
- Most applications have ratings between 4.0 and 4.5, showing positive user feedback.

- Visualization helped in understanding app market trends and user preferences.
