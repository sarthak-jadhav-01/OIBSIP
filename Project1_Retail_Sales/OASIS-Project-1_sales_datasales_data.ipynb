```python
# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load Dataset
df=pd.read_csv("Retail_Sales_Dataset_Project-1.csv")
df.head()
```

```python
# Data Understanding
df.info()
```

```python
# Basic Statistics
df.describe()
```

```python
# Convert Date
df["Date"] = pd.to_datetime(df["Date"])
```

```python
# Total Sales
df["Total Amount"].sum()
```

```python
# Product Category Analysis
df.groupby("Product Category")["Total Amount"].sum()
```

```python
# Bar Graph
df.groupby("Product Category")["Total Amount"].sum().plot(kind="bar")
plt.title("Sales by Product Category")
plt.xlabel("Product Category")
plt.ylabel("Total Sales")
plt.show()
```

```python
# Monthly Sales Analysis 
df["Month"] = df["Date"].dt.month

df.groupby("Month")["Total Amount"].sum().plot(kind="line")
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

```python
# Monthly Sales Analysis
df.groupby("Gender")["Total Amount"].sum().plot(kind="bar")
plt.title("Sales by Gender")
plt.show()
```

```python
df.groupby("Product Category")["Total Amount"].sum().sort_values(ascending=False)
```

```python
df.groupby("Month")["Total Amount"].sum().sort_values(ascending=False)
```

```python
df.groupby("Gender")["Total Amount"].sum()
```

## Conclusion

- Highest sales category: Electronics
- Peak sales month: May
- Gender with more spending: Female
- Sales trend shows variation across months with a peak in mid-year
- Business should focus on Electronics category and target peak sales months to maximize revenue
